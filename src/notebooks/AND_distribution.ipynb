```python
from sklearn.mixture import GaussianMixture
import numpy as np
import matplotlib.pyplot as plt
```

```python
activity_a = lambda n : np.where(np.random.choice(a=[0, 1], size=n, p=[0.5, 0.5]),
         np.random.normal(50, 20, n),
         np.random.normal(200, 20, n)
        )

activity_b = lambda n : np.random.normal(50, 20, n)
```

```python
activity_b(1)
```

```python
and_activities = lambda n : np.max(np.concatenate((
        np.array(activity_a(n), ndmin=2),
        np.array(activity_b(n), ndmin=2)
    )), axis=0)
```

```python
r = and_activities(1000000)
```

```python
plt.hist(r, bins=100 )
```
