```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mu, sd = 10/60, 5/60
n = 100_000

# draw samples from distributions
a = np.random.normal(mu, sd, n)
b = np.random.lognormal(np.log(mu), sd / mu, n)

# bins to use for histogram and x for PDFs
lo, hi = np.min([a, b]), np.max([a, b])
dx = (hi - lo) * 0.06
bins = np.linspace(lo, hi, 101)
x = np.linspace(lo - dx, hi + dx, 501)

# draw figure
fig, [ax1, ax2] = plt.subplots(nrows=2, sharex=True, sharey=True, figsize=(8, 5))

ax1.set_title("Normal draws")
ax1.set_xlim(lo - dx, hi + dx)
ax1.hist(a, bins, density=True, alpha=0.5)
#ax1.plot(x, d1.pdf(x))
#ax1.plot(x, d2.pdf(x), '--')

ax2.set_title("Log-Normal draws")
ax2.hist(b, bins, density=True, alpha=0.5, label="Binned density")
#ax2.plot(x, d1.pdf(x), '--', label="Normal PDF")
#ax2.plot(x, d2.pdf(x), label="Log-Normal PDF")

ax2.legend()
fig.supylabel("Density")
```

```python
from scipy.special import kv

def gig_pdf(x, chi, psi, lambda_):
  """Probability density function of the GIG distribution."""
  return (chi/psi)**(lambda_/2) * x**(lambda_-1) * kv(lambda_-0.5, np.sqrt(chi * x) + np.sqrt(psi/x)) / (2 * np.sqrt(np.pi))

# Parameters
chi = 400
lambda_ = 700
psi = 1000000  # Replace with a finite value if needed

# Generate x values
x = np.linspace(0, 100, 100)  # Adjust range as needed

# Calculate the GIG PDF
pdf = gig_pdf(x, chi, psi, lambda_)
print(pdf)
# Plot the GIG PDF
plt.plot(x, pdf)
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.title('GIG Distribution')
plt.show()
```
